## 1. My lane (or freestyle) and why

**Chosen Lane:** Core Lane 2: Refresh / Content Opportunity Scoring

In search intelligence, a content inventory naturally decays over time due to shifting user intent, search engine updates, or stale information. Reviewing thousands of pages manually is impossible for a standard content operations team. Lane 2 approaches this not as an abstract modeling exercise, but as a practical decision-support optimization problem. By blending an interpretable baseline scoring rule with a machine learning classification model, we can filter out low-volume noise and accurately bubble up the exact pages that have the highest risk of visibility decay but possess the underlying demand to justify a manual rewrite. This lane directly bridges the gap between raw web analytics metrics and scalable, revenue-protecting business actions.

## 2. The question: decision, action, cost of a wrong call

**The Core Question:** Which high-exposure content assets exhibit the highest risk of structural or performance decay, and how can they be prioritized in a ranked queue to maximize manual optimization efficiency?

*   **Unit of Analysis:** A single pseudonymized content item (represented at the `content_hash_id` grain).
*   **The Model Output:** A continuous probability score (0.0 to 1.0) indicating the likelihood that a content asset is experiencing or is about to experience a sustained performance decline, combined with an overall prioritized opportunity score (0 to 100) and automated categorical reason codes (e.g., `stale_visible_page`, `page_one_decay_risk`).
*   **The Actionable Decision:** Content managers and SEO editors will use this ranked queue to decide exactly which pages to check out first on Monday morning. Instead of guessing, they will systematically edit, update, expand, or prune pages starting from the top of the generated list.
*   **The Cost of a Wrong Call:**
    *   *False Positive (Flagging a healthy or unfixable page):* Wastes limited engineering and editorial hours manually reviewing a page that didn't need a refresh, resulting in a high operational opportunity cost.
    *   *False Negative (Missing a genuinely decaying high-traffic page):* Leads to a silent, compounding loss of organic visibility, search impressions, and user sessions, directly hurting business traffic and digital acquisition channels over time.
*   **Why Data and ML Can Help:** Human operators cannot track multi-dimensional trends across thousands of pages simultaneously. While simple rule-based baselines can surface basic stale pages, machine learning models (like Random Forests or Gradient Boosting) can uncover non-linear interactions between complex metrics—such as the subtle inflection points where click-through rate decays even if search position stays flat. Moving from fixed rules to a learned ranking increases Precision@K, ensuring that editorial teams spend their time on pages where data proves an intervention is worth it.
